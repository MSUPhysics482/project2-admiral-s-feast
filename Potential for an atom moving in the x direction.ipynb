```python
import numpy as np                  ## Imports Python's numerical library
import matplotlib.pyplot as plt     ## Imports matplotlib for plotting
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
#arranging a grid for the normalized gaussian to act upon the potential
X = np.arange (-2,2,.01)
Y = np.arange (-2,2,.01)
X, Y = np.meshgrid(X, Y)
#highly simplified, with most values being equal to 1, this is the normalize guassian
G = np.exp(-(X**2+Y**2))/(2*np.pi)
#our potential function, again, with most values being equal to 1 for 
#theoretical simplification
Z = -1*G*(X**2/2+X)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,linewidth=0, antialiased=True)
ax.text2D(0.05, 0.95, "Potential for an atom moving in the x direction with constant velocity", transform=ax.transAxes)
ax.text2D(1, .80, "Potential Strength (arb. units)", transform=ax.transAxes)
ax.set_zlim(-.09, .05)
ax.set_xlabel('\nX axis')
ax.set_ylabel('\nY axis')
ax.set_zlabel('\n \nPotential')
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
