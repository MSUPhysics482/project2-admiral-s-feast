```python
import numpy as np                  ## Imports Python's numerical library
import matplotlib.pyplot as plt     ## Imports matplotlib for plotting
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
#arranging a grid for the normalized gaussian to act upon the potential
X = np.arange (-10,8,.01)
Y = np.arange (-10,8,.01)
X, Y = np.meshgrid(X, Y)
#highly simplified, with most values being equal to 1, this is the normalize guassian
Gx = np.exp(-(Y**2))/(2*np.pi)
Gy = np.exp(-(X**2))/(2*np.pi)
#G= np.exp(-((Y**2)-(X**2)))/(2*np.pi)
#our potential function, again, with most values being equal to 1 for 
#theoretical simplification
V = 1*Gx*(X**2/2)+1*Gy*(Y**2/2+3*Y)
```

Consider an atom of known constant velocity in the +x and +y direction. This atom is placed somewhere along the x-y plane (z=0) inside of a magneto-optical trap. The initial potential describing the interaction of the atom with the electric and magnetic field at any point along the x-y plane is displayed below.

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, V,cmap=cm.coolwarm,linewidth=0, antialiased=True)
#ax.text2D(0.05, 0.95, transform=ax.transAxes)
ax.text2D(1, .80, "Potential Strength (arb. units)", transform=ax.transAxes)
ax.set_zlim(0, 4)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potential (arb. units)')
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
fig, ax = plt.subplots()
cs = ax.contourf(X, Y, V, cmap=cm.coolwarm)
cbar = fig.colorbar(cs)
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

These plots demonstrate a number of renewed interests about the potential. We see that the minimum strength of the potential is shifted slightly off x = 0 and y = 0. This potential well is a result of the doppler shift in both directions. The atom moving with velocity in the positive xy direction will see the laser doppler shifted into resonance in positive xy, moving the repulsive potential closer to the origin. The laser is doppler shifted out of resonance in negative xy, moving the repulsive potential away from the origin. The laser beam profile is evident when observing strength in steps along the xy-axes. We see a reduced strength away from the central beam axis.
